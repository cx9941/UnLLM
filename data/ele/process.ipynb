```python
import pandas as pd
import json
data = json.load(open('raw_data/50EleReviews.json', 'r'))
df = pd.DataFrame({'X': data['X'], 'y':data['y']})
df['y'] = df['y'].apply(lambda x: data['target_names'][x])
df.columns = ['text', 'label']
df = df.drop_duplicates()
df.to_csv('raw_data/50EleReviews.csv', index=None)
```
