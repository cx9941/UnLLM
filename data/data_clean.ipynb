```python
import pandas as pd
import os
dataset_list = ['ele', 'news', 'thucnews']
# dataset_list = ['banking', 'clinc', 'stackoverflow']
ratio_list = [0.25, 0.5 ,0.75]
for dataset in dataset_list:
    for ratio in ratio_list:
        df = pd.read_csv(f'{dataset}/{dataset}_{ratio}/idx.txt', sep='\t')
        df = df.sort_values('index')[['node']]
        df['node'] = df['node'].apply(lambda x: x.replace(' ', '_'))
        df.to_csv(f"{dataset}/{dataset}_{ratio}.txt", index=None, header=None)
        os.system(f"rm -rf {dataset}/{dataset}_{ratio}")
        os.system(f"rm -rf {dataset}/origin_data/candidate_class.tsv")
        os.system(f"rm -rf {dataset}/origin_data/prob.tsv")
```
