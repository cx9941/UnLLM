```python
import pandas as pd
import json
ans = {}
pd_ans = {'dataset': [], 'ratio':[], 'split':[], 'value':[]}
for dataset_name in ['demo', 'banking', 'clinc', 'stackoverflow', 'ele', 'news', 'textures']:
    ans[dataset_name] = {}
    for rate in [0.25, 0.5, 0.75]:
        ans[dataset_name][rate] = {}
        label_ratio = pd.read_csv(f'{dataset_name}/{dataset_name}_{rate}.txt', header=None)[0].tolist()
        for split in ['train', 'eval', 'test']:
            df = pd.read_csv(f"{dataset_name}/origin_data/{split}.tsv", sep='\t')
            sub_df = df[df['label'].isin(label_ratio)]
            ans[dataset_name][rate][split] = len(sub_df)
            pd_ans['dataset'].append(dataset_name)
            pd_ans['split'].append(split)
            pd_ans['ratio'].append(rate)
            pd_ans['value'].append(len(df) if split == 'test' else len(sub_df))
json.dump(ans, open('statics/sample_num.json', 'w'))
```

```python
pd_ans = pd.DataFrame(pd_ans).pivot(index='dataset', columns=['ratio', 'split'], values=['value'])
pd_ans = pd_ans.applymap(lambda x: "{:,}".format(x))
pd_ans.to_excel('statics/data_statics.xlsx')
pd_ans
```

```python
data_statics = pd_ans.to_latex().replace(' ', '').replace('0000', '')
replace_map = {
    "Classes": '|Classes|',
    "train": '|Train|',
    "eval": '|Validation|',
    "test": '|Test|',
    "banking": 'BANKING',
    "clinc": 'CLINC',
    "stackoverflow": 'STACK',
    "thucnews": 'CNEWS',
    "atis": 'ATIS',
    "snips": 'SNIPS',
    "ele": 'REVIEWS',
    "news": 'NEWSG',
    "0000": "",
    ".00": "",
    "lrrrrrrr": "l|p{1.2cm}p{1.2cm}p{1.2cm}p{1.2cm}p{1.2cm}p{1.2cm}p{1.2cm}",
    "toprule\n": "toprule\nDataset"
}
for i, v in replace_map.items():
    data_statics = data_statics.replace(i, v)
print(data_statics.replace(' ', ''))
```

```python
import pandas as pd
ans = {}
pd_ans = {'dataset': [], 'ratio':[], 'split':[], 'value':[]}
for dataset_name in ['banking', 'clinc', 'stackoverflow', 'thucnews', 'atis', 'ele', 'news','mcid', 'snips']:
    ans[dataset_name] = {}
    ans[dataset_name]['Classes'] = len(pd.read_csv(f"{dataset_name}/origin_data/label.list", header=None))
    all_df = []
    for split in ['train', 'eval','test']:
        df = pd.read_csv(f"{dataset_name}/origin_data/{split}.tsv", sep='\t')
        all_df.append(df)
        ans[dataset_name][f"{split}"] = int(len(df))
    all_df = pd.concat(all_df)
    all_df['len'] = all_df['text'].apply(lambda x: len(x.split(' ')))
    ans[dataset_name]['Length (Avg.)'] = all_df['len'].mean().round(2)

ans = pd.DataFrame(ans)
ans = ans.T
for col in ans.columns:
    if 'Length' not in col:
        ans[col] = ans[col].apply(lambda x: int(x))
ans
```
